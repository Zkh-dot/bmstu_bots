```python
import requests
from bs4 import BeautifulSoup
import hashlib
import re

hash = []
res = requests.post('https://e-learning.bmstu.ru/iu5/mod/folder/view.php?id=1630', data={'page': '3'})
# html = pandas.read_html(res.content)
```

```python
soup = BeautifulSoup(res.text)

with open("soup_file.html", "w") as file:
    file.write(soup.prettify())
```

```python
BUF_SIZE = 65536
md5 = hashlib.md5()

with open("soup_file.html", 'rb') as f:
    while True:
        data = f.read(BUF_SIZE)
        if not data:
            break
        md5.update(data)

hash.append(md5.hexdigest())
```
